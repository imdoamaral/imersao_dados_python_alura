```python
# Create a dictionary for translating 'senioridade' categories
senioridade_translation = {
    'SE': 'Senior',
    'MI': 'Médio',
    'EN': 'Iniciante',
    'EX': 'Executivo'
}

# Replace the values in the 'senioridade' column
df['senioridade'].replace(senioridade_translation, inplace=True)

# Display the updated value counts
print("Categorias da coluna 'senioridade' traduzidas:")
display(df['senioridade'].value_counts())
```

```python
# Create a dictionary for translating 'contrato' categories
contrato_translation = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

# Replace the values in the 'contrato' column
df['contrato'].replace(contrato_translation, inplace=True)

# Display the updated value counts
print("Categorias da coluna 'contrato' traduzidas:")
display(df['contrato'].value_counts())
```

```python
# Create a dictionary for translating 'tamanho_empresa' categories
tamanho_empresa_translation = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}

# Replace the values in the 'tamanho_empresa' column
df['tamanho_empresa'].replace(tamanho_empresa_translation, inplace=True)

# Display the updated value counts
print("Categorias da coluna 'tamanho_empresa' traduzidas:")
display(df['tamanho_empresa'].value_counts())
```

```python
# Create a dictionary for translating 'remoto' categories
remoto_translation = {
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
}

# Replace the values in the 'remoto' column
df['remoto'].replace(remoto_translation, inplace=True)

# Display the updated value counts
print("Categorias da coluna 'remoto' traduzidas:")
display(df['remoto'].value_counts())
```

```python
import numpy as np

# Criação de um dataframe de teste
df_salarios = pd.DataFrame({
    'nome': ['Israel', 'Matias', 'Amaral', 'Israelzinho', 'Israelzao'],
    'salario': [4000, np.nan, 5000, np.nan, 100000],
})

# Cria uma coluna de 'salario_media', substitui os nulos pela media e arredonda
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

# Cria uma coluna de 'salario_mediana' e substitui os nulos pela mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

# ffill = forward fill = preenche com o valor anterior
df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()

df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

# bfill = back fill = preenche com o valor posterior
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()

df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Israel', 'Matias', 'Amaral', 'Israelzinho', 'Israelzao'],
    'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belém']
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')

df_cidades
```
