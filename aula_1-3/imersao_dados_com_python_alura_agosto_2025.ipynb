## Tradução das colunas para pt-BR via Gemini

```python
df.columns
```

```python
# Create a dictionary for translating column names
translation_dict = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

# Rename the columns using the dictionary
df.rename(columns=translation_dict, inplace=True)

# Display the translated columns
print("Colunas traduzidas:")
print(df.columns)
```

```python
# Create a dictionary for translating 'senioridade' categories
senioridade_translation = {
    'SE': 'senior',
    'MI': 'pleno',
    'EN': 'junior',
    'EX': 'executivo'
}

# Replace the values in the 'senioridade' column
df['senioridade'].replace(senioridade_translation, inplace=True)

# Display the updated value counts
print("Categorias da coluna 'senioridade' traduzidas:")
display(df['senioridade'].value_counts())
```

```python
# Create a dictionary for translating 'contrato' categories
contrato_translation = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

# Replace the values in the 'contrato' column
df['contrato'].replace(contrato_translation, inplace=True)

# Display the updated value counts
print("Categorias da coluna 'contrato' traduzidas:")
display(df['contrato'].value_counts())
```

```python
# Create a dictionary for translating 'tamanho_empresa' categories
tamanho_empresa_translation = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}

# Replace the values in the 'tamanho_empresa' column
df['tamanho_empresa'].replace(tamanho_empresa_translation, inplace=True)

# Display the updated value counts
print("Categorias da coluna 'tamanho_empresa' traduzidas:")
display(df['tamanho_empresa'].value_counts())
```

```python
# Create a dictionary for translating 'remoto' categories
remoto_translation = {
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
}

# Replace the values in the 'remoto' column
df['remoto'].replace(remoto_translation, inplace=True)

# Display the updated value counts
print("Categorias da coluna 'remoto' traduzidas:")
display(df['remoto'].value_counts())
```

```python
import numpy as np

# Criação de um dataframe de teste
df_salarios = pd.DataFrame({
    'nome': ['Israel', 'Matias', 'Amaral', 'Israelzinho', 'Israelzao'],
    'salario': [4000, np.nan, 5000, np.nan, 100000],
})

# Cria uma coluna de 'salario_media', substitui os nulos pela media e arredonda
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

# Cria uma coluna de 'salario_mediana' e substitui os nulos pela mediana
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

# ffill = forward fill = preenche com o valor anterior
df_temperaturas['preenchido_ffill'] = df_temperaturas['temperatura'].ffill()

df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'dia': ['segunda', 'terça', 'quarta', 'quinta', 'sexta'],
    'temperatura': [30, np.nan, np.nan, 28, 27]
})

# bfill = back fill = preenche com o valor posterior
df_temperaturas['preenchido_bfill'] = df_temperaturas['temperatura'].bfill()

df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome': ['Israel', 'Matias', 'Amaral', 'Israelzinho', 'Israelzao'],
    'cidade': ['São Paulo', np.nan, 'Curitiba', np.nan, 'Belém']
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('Não informado')

df_cidades
```

## Desafio: salário por país para Cientistas de Dados

```python
# Desafio: salário por país para Cientistas de Dados

# Filtrar o DataFrame para o cargo de 'Data Scientist'
df_data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist'].copy()

# Calcular o salário médio por país para Cientistas de Dados e classificar em ordem decrescente
salarios_pais_ds = df_data_scientist.groupby('residencia')['usd'].mean().sort_values(ascending=False).reset_index()

# Limitar a quantidade de barras exibidas (por exemplo, os top 10)
top_n = 10
salarios_pais_ds_top_n = salarios_pais_ds.head(top_n)

# Criar um gráfico de barras usando Plotly Express
fig = px.bar(salarios_pais_ds_top_n,
           x='residencia',
           y='usd',
           title=f'Top {top_n} Países com Maior Salário Médio para Cientistas de Dados (USD)',
           labels={'residencia': 'País de Residência', 'usd': 'Salário Médio Anual (USD)'})

# Exibir o gráfico
fig.show()
```

```python
import pycountry

# Funçao para converter de ISO-2 para ISO-3
def iso2_to_iso3(iso2):
  try:
    # Corrected the attribute name from alhpa_2 to alpha_2
    return pycountry.countries.get(alpha_2=iso2).alpha_3
  except:
    return None

# Cria nova coluna com codigo ISO-3
df_limpo['residencia_iso3'] = df_limpo['residencia'].apply(iso2_to_iso3)

# Calcula a media salarial por país (ISO-3)
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']
media_ds_pais = df_ds.groupby('residencia_iso3')['usd'].mean().reset_index()

# Gerar o mapa
fig = px.choropleth(media_ds_pais,
                    locations='residencia_iso3',
                    color='usd',
                    color_continuous_scale='rdylgn',
                    title='Salário médio de Cientista de Dados por país',
                    labels={'usd': 'Salário médio (USD)', 'residencia_iso3': 'País'}
                    )

fig.show()
```
